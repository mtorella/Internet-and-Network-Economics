# Sector Extraction and Analysis
Since the sector mapping process is crucial in our analysis but also very complicated, this Jupyter notebook is dedicated to extracting and analyzing sectors from raw dataset growth account and intangibles analytical CSV files. We used this file since the provided outputs are more easily interpretable and can be directly used for further analysis.

```python
import pandas as pd
import os

# Set up paths
data_raw_path = os.path.join('..', 'data', 'raw')
print(f"Data raw path: {data_raw_path}")
```

## 2. Load the Raw Dataset
Load the growth account CSV file into a pandas DataFrame and display basic information about the dataset structure.

```python
# Load growth accounts CSV
growth_accounts_path = os.path.join(data_raw_path, 'growth accounts.csv')
df_growth = pd.read_csv(growth_accounts_path)

print("Growth Accounts Dataset Shape:", df_growth.shape)
print("\nFirst few rows:")
df_growth.head()

# Unique NACE codes in growth accounts
unique_nace = df_growth['nace_r2_code'].unique()
print("\nUnique NACE codes in growth accounts:", unique_nace)
```

## 4. Extract Unique Intangibles Analytical Sectors
Load the intangibles analytical CSV file and extract unique sectors from it.

```python
# Load intangibles analytical CSV
intangibles_path = os.path.join(data_raw_path, 'intangibles analytical.csv')
df_intangibles = pd.read_csv(intangibles_path)

print("Intangibles Analytical Dataset Shape:", df_intangibles.shape)
print("\nFirst few rows:")
df_intangibles.head()

# Unique NACE codes in intangibles analytical
unique_nace_intangibles = df_intangibles['nace_r2_code'].unique()
print("\nUnique NACE codes in intangibles analytical:", unique_nace_intangibles)
```
